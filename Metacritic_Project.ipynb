```python
#Installing libraries 
%pip install re
%pip install requests
%pip install pandas

#Importing libraries
import re
import requests
import pandas as pd
from IPython.display import HTML

#Solving the URL Redirect issues
user_agents = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}

#Fetching URL from metacritic
contact = requests.get('https://www.metacritic.com/browse/movie/all/all/2013/metascore/?page=1', headers=user_agents)

print(re)

#Use regex for information
#Extract title
movie_Title = re.findall(r'<div data-title="(.*?)"', contact.text)
#Extract release date
movie_Release_Date = re.findall(r'<span class="u-text-uppercase">\s+(.*)\s+<\/span>', contact.text)
#Extract metascore
metascore = re.findall(r'<div title="Metascore (.*?) out of 100"', contact.text)
#Extract thumbnail
thumbnail = re.findall(r'<img src="(.*?)" height="', contact.text)

fixed = []
for url in thumbnail:
    url = url.replace("&amp;", "&")
    fixed.append(url)
#Extract summary
summary = re.findall(r'<div class="c-finderProductCard_description"><span>(.*?)<\/span><\/div> <div class="c-finderProductCard_meta g-outer-spacing-top-auto">', contact.text)

#Create data frame for the extracted data
import pandas as pd
data = {'Movie Title': movie_Title,
        'Movie Release Date': movie_Release_Date,
        'Metascore': metascore,
        'Thumbnail': fixed,
        'Summary': summary}
  
moviesdf = pd.DataFrame(data)

HTML(moviesdf.to_html(render_links=True, escape=False))

```
